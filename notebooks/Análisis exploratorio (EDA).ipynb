## Usando pandas, seaborn y matplotlib:
    Visualizaciones obligatorias:
    • Distribución de ventas (histograma + boxplot).
    • Ventas por región y categoría.
    • Top 10 productos y clientes.
    • Tendencia mensual de ventas.
    • Heatmap región vs categoría.
    Cada grafico debe incluir una interpretación (en ingles).

```python
# 1 Importaciones y carga de datos
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración visual
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# Cargar datos 
df = pd.read_csv("Datos_limpio.csv")

# Verificación rápida
print(df.head())
print(df.info())
```

## Distribución de ventas (Histogram + Boxplot)

```python
# HISTOGRAMA

plt.figure()
sns.histplot(df["total"], bins=30, kde=True)
plt.title("Distribution of Total Sales")
plt.xlabel("Total Sale Amount")
plt.ylabel("Frequency")
plt.show()
```

### The chart indicates that most sales are concentrated in the low-to-mid range, while a few high-value transactions account for significantly higher amounts.

```python
# Boxplot

plt.figure()
sns.boxplot(x=df["total"])
plt.title("Boxplot of Total Sales")
plt.xlabel("Total Sale Amount")
plt.show()
```

### Most sales fall between 10,000 and 30,000, although there are some high-value outliers exceeding 50,000.

## Ventas por región y categoría

```python
# región
ventas_region = df.groupby("pais")["total"].sum().reset_index()

plt.figure()
sns.barplot(data=ventas_region, x="pais", y="total")
plt.title("Total Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

### All three countries show nearly identical sales levels, indicating a balanced market performance across the region.

```python
# Categoría
ventas_categoria = df.groupby("tipo_producto")["total"].sum().reset_index()

plt.figure()
sns.barplot(data=ventas_categoria, x="tipo_producto", y="total")
plt.title("Total Sales by Product Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)
plt.show()
```

### Sales are consistent across all categories, reflecting balanced market demand.

## Top 10 productos y clientes

```python
# Top 10 productos

top_productos = (
    df.groupby("producto")["total"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure()
sns.barplot(data=top_productos, x="total", y="producto")
plt.title("Top 10 Products by Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Product")
plt.show()
```

```python
# top 10 clientes
top_clientes = (
    df.groupby("segmento_cliente")["total"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
    .reset_index()
)

plt.figure()
sns.barplot(data=top_clientes, x="total", y="segmento_cliente")
plt.title("Top 10 Customers by Total Sales")
plt.xlabel("Total Sales")
plt.ylabel("Customer Segment")
plt.show()
```

### Sales are uniform across the Top 10 Products, while the Corporate segment stands out as the main engine for business growth in the Top 10 Clients list.

## Tendencia mensual de ventas

```python
df["fecha"] = pd.to_datetime(df["fecha"])

ventas_mensuales = (
    df.groupby(pd.Grouper(key="fecha", freq="M"))["total"]
    .sum()
    .reset_index()
)

plt.figure()
sns.lineplot(data=ventas_mensuales, x="fecha", y="total", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.show()
```

### Sales have plummeted since the beginning of January, dropping from 3.0 units to less than 0.5 by February 2026.

## Heatmap región vs categoría

```python
heatmap_data = pd.pivot_table(
    df,
    values="total",
    index="pais",
    columns="tipo_producto",
    aggfunc="sum"
)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="Blues")
plt.title("Sales Heatmap: Region vs Product Category")
plt.xlabel("Product Category")
plt.ylabel("Region")
plt.show()
```

### Regional analysis reveals shifting consumer preferences, with distinct product categories dominating in each specific country
